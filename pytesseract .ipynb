```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('hit.jpg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```
